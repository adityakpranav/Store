## Data Preparation : Data Cleaning

### Duplicate values

```python
Data_df_unique = Data_df[~Data_df.duplicated()]
print("No Of Rows in Original dataset: %s"%Data_df.shape[0])
print("No Of Rows in Unique dataset: %s"%Data_df_unique.shape[0])
```

### Variable type

```python
Data_df_unique.dtypes[Data_df_unique.dtypes=='object']
```

```python

schema = {"Method":'category',
          "Type":'category',
          "Date":'datetime64'
         
         }
Data_df_unique = Data_df_unique.astype(schema) 
print(Data_df_unique.dtypes)
print("----\n For how many attributes did you need to change the data type? %s"%len(schema))
print(Data_df_unique.dtypes)
```

### Missing value treatment

```python
Data_df_unique.isna().sum()
```

```python
Data_df_unique.columns[Data_df_unique.isnull().sum()>0]
```

```python
Data_df_unique_fillna = Data_df_unique.fillna( method ='ffill').fillna( method ='bfill')
```

## Outlier Treatment: 

### Identify the varibales

```python
Data_df_unique_fillna_Numeric = Data_df_unique_fillna[Data_df_unique_fillna._get_numeric_data().columns]
Data_df_unique_fillna_Numeric
```

### Outliers

```python
Data_df_unique_fillna_Numeric.describe(percentiles  =[.20,.80])
Q1 = Data_df_unique_fillna_Numeric.quantile(0.25)
Q3 = Data_df_unique_fillna_Numeric.quantile(0.75)
IQR = Q3 - Q1

Data_df_unique_fillna_Numeric_outlier = Data_df_unique_fillna_Numeric[~((Data_df_unique_fillna_Numeric < (Q1 - 1.5 * IQR)) |(Data_df_unique_fillna_Numeric > (Q3 + 1.5 * IQR))).any(axis=1)]
Data_df_unique_fillna_Numeric_outlier.describe()
```

## Data Preparation Feature Engineering

### Feature Transformation

```python
from sklearn.linear_model import LinearRegression
scores = {}
y = Data_df_unique_fillna_Numeric_outlier["Price"]
for col in Data_df_unique_fillna_Numeric_outlier.columns:
    X = np.array(Data_df_unique_fillna_Numeric_outlier[col]).reshape(-1,1)
    reg = LinearRegression().fit(X, y)
    scores[col]=reg.score(X, y)

print(scores)
scores_df = pd.DataFrame(scores.items(),columns=['attribute','Score'])
scores_df.sort_values(by=['Score'])
```

```python
print("Identify variables that have non-linear trends \n %s"%Data_df_unique_fillna_Numeric_outlier.columns[Data_df_unique_fillna_Numeric_outlier.columns!='Price'])
```
